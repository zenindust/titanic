# Lab 4 Titanic


```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```
# Upload file titanic_disaster.csv lên Colab trước
train = pd.read_csv('/content/titanic_disaster.csv')

print(train.shape)
train.head()
```

```
print(train.info())
print(train.isnull().sum())
```

```
plt.figure(figsize=(10,5))
sns.heatmap(train.isnull(), cbar=False)
plt.title('Missing Values')
plt.show()
```

```
train[['secondName','firstName']] = train['Name'].str.split(',', expand=True)
train.drop('Name', axis=1, inplace=True)

train['Sex'] = train['Sex'].replace({
    'male':'M',
    'female':'F'
})

train.head()
```

```
sns.boxplot(x='Pclass', y='Age', data=train)
plt.show()

train['Age'] = train.groupby('Pclass')['Age'].transform(
    lambda x: x.fillna(x.mean())
)
```

```
def age_group(age):
    if age <= 12:
        return 'Kid'
    elif age <= 18:
        return 'Teen'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Older'

train['AgeGroup'] = train['Age'].apply(age_group)
```

```
train['namePrefix'] = train['secondName'].str.extract('([A-Za-z]+)\.')
train['familySize'] = 1 + train['SibSp'] + train['Parch']
train['Alone'] = np.where(train['familySize']==1,1,0)

train['Cabin'] = train['Cabin'].fillna('Unknown')
train['typeCabin'] = train['Cabin'].astype(str).str[0]

train.head()
```

```
sns.countplot(x='Sex', hue='Survived', data=train)
plt.show()
```

```
sns.countplot(x='Pclass', hue='Survived', data=train)
plt.show()
```

```
sns.catplot(
    x='AgeGroup',
    hue='Survived',
    col='Sex',
    kind='count',
    data=train
)
plt.show()
```

```
sns.barplot(
    x='familySize',
    y='Survived',
    data=train
)
plt.show()
```

```
sns.boxplot(
    x='Survived',
    y='Fare',
    data=train
)
plt.show()
```

```
sns.catplot(
    x='Pclass',
    hue='Survived',
    col='Embarked',
    kind='count',
    data=train
)
plt.show()
```

```
train.to_csv('Titanic_processed.csv', index=False)
print('Đã lưu Titanic_processed.csv')
```
